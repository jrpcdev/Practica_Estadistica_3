```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
import math
import seaborn as sns
```

```python
df1=pd.read_excel("Clima_mensual.xlsx")
df2=pd.read_excel("mat5.xlsx")
```

```python
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, on=['Año','Mes'], how='inner')
```

```python
df.head()
```

```python
df = df.rename(columns={"muertos_en_accidentes_de_trafico_en_los_ultimos_10_años (1)": "muertes"})
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
columna_seleccionada = 'Num_dias_lluvia'
plt.figure(figsize=(12, 8))
for year, data_year in df.groupby('Año'):
    sns.kdeplot(data=data_year[columna_seleccionada], label=str(year), fill=True)

plt.title(f'Distribución de nº de dias de lluvia por Año')
plt.xlabel(columna_seleccionada)
plt.ylabel('Densidad de Probabilidad')
plt.legend(title='Año')
plt.show()
```

```python
columna_seleccionada = 'Num_dias_lluvia'
plt.figure(figsize=(12, 8))
for month, data_month in df.groupby('Mes'):
    sns.kdeplot(data=data_month[columna_seleccionada], label=str(month), fill=True)

plt.title(f'Distribución de nº de dias de lluvia por mes')
plt.xlabel(columna_seleccionada)
plt.ylabel('Densidad de Probabilidad')
plt.legend(title='Mes')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load Files
folder = 'data'
serieTemporalBase = f'{folder}/Clima_y_Accidentes.xlsx'


dfserieTemporalBase = pd.read_excel(serieTemporalBase)
dfserieTemporalBase['cuartil_lluvia'] = pd.qcut(dfserieTemporalBase['Num_dias_lluvia'], 2, labels=False)
dfserieTemporalBase['cuartil_niebla'] = pd.qcut(dfserieTemporalBase['Num_dias_niebla'], 2, labels=False)


dfserieTemporalBase = dfserieTemporalBase.iloc[:,[True,False, False, False,False,False,True,True,True, True]]
dfserieTemporalBase = dfserieTemporalBase.groupby(['MitadAnio', 'cuartil_lluvia', 'cuartil_niebla', 'AÃ±o']).sum()
dfserieTemporalBase = dfserieTemporalBase.groupby(['MitadAnio', 'cuartil_lluvia', 'cuartil_niebla']).mean()

dfserieTemporalBase['p'] = dfserieTemporalBase['Accidentes'] / 1416.297222
dfserieTemporalBase
```

```python
nodo0 = dfserieTemporalBase.groupby('MitadAnio').sum()
nodo0['H'] = -nodo0['p']*np.log2(nodo0['p'])-(1 - nodo0['p'])*np.log2(1 - nodo0['p'])
nodo0
```
